# **Web Scraping** 
    - Scraping data indomaret dan alfamaret daerah jember dan jaktim
    - Link: https://www.tiendeo.co.id/toko/jakarta-timur/indomaret 
    - Extension: selector gadget 

```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup
from bs4.element import Comment
```

#### **Indomaret**

```python
df1 = []
for page_number in range(1, 4):
    page_response = requests.get(f'https://www.tiendeo.co.id/toko/jakarta-timur/indomaret')
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    nama_tokos = page_soup.select('#nearest-stores-slider .typo-subhead-14-med')
    for nama_toko in nama_tokos:
        row = nama_toko
        df1.append(row)
    data1 = pd.DataFrame(df1, columns=["NAMA_TOKO"]) 
```

```python
df2 = []
for page_number in range(1, 4):
    page_response = requests.get(f'https://www.tiendeo.co.id/toko/jakarta-timur/indomaret')
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    alamat_tokos = page_soup.select('#nearest-stores-slider .typo-body-13-reg')
    for alamat_toko in alamat_tokos:
        row = alamat_toko
        df2.append(row)
    data2 = pd.DataFrame(df2, columns=["ALAMAT"]) 
```

```python
frame = [data1,data2]
result = pd.concat(frame,axis=1)
result['City'] = "Jakarta Timur"
result = result.drop_duplicates()
```
