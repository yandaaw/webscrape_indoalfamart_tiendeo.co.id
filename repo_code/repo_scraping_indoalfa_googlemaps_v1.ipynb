```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup
from bs4.element import Comment
```

# **Indomaret**

```python
df1 = []
for page_number in range(1, 4):
    page_response = requests.get(f'https://www.loveindonesia.com/directory/id/jakarta/search/business=indomaret%7Clocation=Kota+Jakarta+Timur%2C+Jakarta/0/{page_number}')
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    nama_tokos = page_soup.select('.title a')
    for nama_toko in nama_tokos:
        row = nama_toko
        df1.append(row)
    data1 = pd.DataFrame(df1, columns=["NAMA_TOKO"]) 
```

```python
df2 = []
for page_number in range(1, 4):
    page_response = requests.get(f'https://www.loveindonesia.com/directory/id/jakarta/search/business=indomaret%7Clocation=Kota+Jakarta+Timur%2C+Jakarta/0/{page_number}')
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    alamat_tokos = page_soup.select('.col_list_desc div:nth-child(1) .icon_16+ div')
    for alamat_toko in alamat_tokos:
        row = alamat_toko
        df2.append(row)
    data2 = pd.DataFrame(df2, columns=["ALAMAT"]) 
```

```python
# Merge
frame = [data1,data2]
result = pd.concat(frame,axis=1)

# Add City
result['City'] = "Jakarta Timur"

# Drop duplicate
result = result.drop_duplicates()
```

```python
# Drop un-indomaret
result = result.drop([result.index[27]])
```
